```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

**Loaded the CSV file of crop yield**

```python
df = pd.read_csv("crop_yield_sorted.csv")
df
```

```python
df.nunique()
```

*Sorted the data by year wise and state wise*

```python
df = df.sort_values(by = ['Crop_Year', 'State'])
df = df.reset_index(drop=True)
```

**Extracted the data of UP**

```python
#dfup is for Uttar Pradesh data , Here w is for Wheat and r for rice
dfup=df[(df['State'] == 'Uttar Pradesh')].reset_index(drop=True)
dfupw=df[(df['State'] == 'Uttar Pradesh') & (df['Crop']=='Wheat')].reset_index(drop=True)
dfupr= df[((df['State'] == 'Uttar Pradesh') & (df['Crop']=='Rice'))].reset_index(drop=True)
x1 = dfupw['Crop_Year']
x2 = dfupr['Crop_Year']
y1 = dfupw['Fertilizer']
y2 = dfupr[dfupr['Season']=='Kharif']
dfupr
```
